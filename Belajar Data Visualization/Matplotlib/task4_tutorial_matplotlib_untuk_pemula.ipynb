## Membangun Multi Panel (Tugas: Membangun Dashboard Analisis Saham 2-Panel)
Gunakan kode preparasi data yang sama seperti sebelumnya, lalu bangun visualisasi dengan spesifikasi ketat berikut:

1. Inisiasi Multi-Axes: Buat sebuah figure yang langsung memiliki 2 area grafik (axes) yang ditumpuk secara vertikal (2 baris, 1 kolom).

    - Petunjuk Logika: Gunakan parameter nrows dan ncols pada plt.subplots(). Ingat, fungsi ini sekarang akan mengembalikan array berisi dua objek ax, bukan lagi satu. Kamu bisa menangkapnya dengan variabel (ax1, ax2).

2. Sinkronisasi Sumbu X: Karena kedua grafik menggunakan garis waktu (tanggal) yang persis sama, sinkronkan sumbu X keduanya. Jika nanti grafiknya di-zoom, keduanya harus bergerak bersamaan.

    - Petunjuk Logika: Cari dan gunakan argumen sharex=True saat inisiasi subplot.

3. Panel Atas (ax1):

    - Plot kolom Close sebagai line chart (garis biasa).

    - Beri warna biru.

    - Beri judul panel: "Tren Harga Penutupan".

4. Panel Bawah (ax2):

    - Plot kolom Volume. Alih-alih menggunakan garis, gunakan Bar Chart (diagram batang) karena volume merepresentasikan kuantitas diskrit per hari.

    - Petunjuk Logika: Gunakan metode ax2.bar().

    - Beri warna abu-abu (color='gray').

    - Beri judul panel: "Volume Transaksi Harian".

5. Finishing:

    - Terapkan fig.autofmt_xdate() dan fig.tight_layout() agar marginnya rapi.

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
# import datanya
df = pd.read_csv('FBCopy.csv')

# Mengonversi string menjadi format waktu dan menjadikannya indeks
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date').sort_index()

cols_close = df['Close']
cols_volume = df['Volume']

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(cols_close, color='blue')
ax1.set_title('Tren Harga Penutupan')

ax2.bar(df.index, cols_volume, color='gray')
ax2.set_title('Volume Transaksi Harian')
ax2.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

fig.autofmt_xdate()
fig.tight_layout()
plt.show()
```
