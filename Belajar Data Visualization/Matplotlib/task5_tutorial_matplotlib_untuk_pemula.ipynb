## Berikut adalah spesifikasi tugas analitismu selanjutnya:

1. Inisiasi Tunggal: Buat kanvas Object-Oriented dengan satu area grafik (fig, ax).

2. Grafik Pencar (Scatter Plot): Tinggalkan line chart sejenak. Gunakan metode ax.scatter() untuk memetakan hubungan dua variabel. Letakkan Volume di sumbu X dan Close di sumbu Y.

3. Penanganan Kepadatan Data: Tambahkan argumen alpha=0.5 pada plot tersebut. Parameter transparansi ini wajib digunakan dalam visualisasi berskala besar agar kita bisa membedakan secara visual area mana yang titik datanya padat menumpuk dan area mana yang merupakan outliers (pencilan).

4. Standardisasi Anatomi: Beri judul dan label sumbu yang spesifik. Terapkan kembali teknik StrMethodFormatter dari modul ticker, tetapi kali ini aplikasikan pada sumbu X agar besaran volume transaksi mudah dibaca dan tidak menjadi notasi ilmiah.

5. Analisis Diagnostik: Setelah grafiknya berhasil dirender, jangan hanya mengirimkan susunan kodenya. Sertakan satu paragraf observasi analitismu. Jelaskan apakah sebaran titiknya membentuk pola garis (korelasi linear), mengelompok di satu area tertentu, atau justru menyebar acak sepenuhnya tanpa pola matematis yang jelas.

```python
%matplotlib widget
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
import numpy as np

df = pd.read_csv('FBCopy.csv')

df = df.set_index('Volume').sort_index()

cols_close = df['Close']

fig, ax = plt.subplots()
ax.plot(cols_close, color='blue', alpha=0.5)
ax.set_title('Korelasi Harga Penutupan dengan Besara Volume')
ax.set_ylabel('Harga Penutupan')
ax.set_xlabel('Volume')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
fig.autofmt_xdate()
fig.tight_layout()
plt.show()
```
