```python
import matplotlib.pyplot as plt
import numpy as np
```

Pembuatan Data: Buat variabel berisi array [0,1,2,...,9] sebagai data dummy.

```python
dataDummy = np.arange(10)
```

Interaktivitas: Tuliskan magic command yang digunakan untuk mengaktifkan mode interaktif pada grafik sehingga nantinya grafik bisa di-zoom atau digeser (pan).

```python
%matplotlib widget
```

Visualisasi: Buatlah sebuah line chart yang paling sederhana dari data yang sudah dibuat.

```python
fig, ax = plt.subplots()
plt.plot(dataDummy)
plt.show()
```

Penyimpanan (Eksperimen): Tuliskan kode untuk menyimpan grafik tersebut ke dalam file berekstensi PNG dengan syarat khusus: grafiknya harus versi resolusi tinggi dan pastikan whitespace di pinggir grafik dihilangkan (dipotong secara ketat).

**Kesalahan** = 
1. Seharusnya disimpan terlebih dahulu baru di show()
2. Menggunakan plot dengan ax bukan plt lagi

`Kode Seharusnya`:

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Persiapan Data & Interaktivitas
%matplotlib widget
dataDummy = np.arange(10)

# 2. Inisialisasi Kanvas (Pendekatan Object-Oriented yang konsisten)
fig, ax = plt.subplots()

# 3. Plotting menggunakan metode yang terikat langsung pada objek axes
ax.plot(dataDummy)

# 4. Eksekusi penyimpanan ke file dilakukan TERLEBIH DAHULU
# Perhatikan juga penambahan ekstensi .png pada nama file agar formatnya jelas
fig.savefig('task_pengantar_matplotlib.png', 
            dpi=400,
            bbox_inches='tight')

# 5. Tampilkan ke layar sebagai proses terminasi blok visualisasi
plt.show()
```
