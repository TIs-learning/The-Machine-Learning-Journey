## Inisialisasi Kanvas: 
1. Buat sebuah objek kanvas kosong (figure).

2. Pembuatan Subplot: Bagi kanvas tersebut menjadi grid berukuran 2x2, lalu tambahkan tiga subplot yang menempati posisi kiri atas, kanan atas, dan kiri bawah.

3. Simulasi Data: Buat sebuah data dummy yang mensimulasikan pergerakan acak (random walk, menyerupai fluktuasi harga saham) dengan menggenerasi 50 angka acak berdistribusi normal, lalu akumulasikan nilainya menggunakan fungsi kumulatif sum.

4. Visualisasi Terpisah & Styling: Plot data yang sama tersebut ke dalam tiga subplot yang sudah kamu buat dengan gaya berikut:

    - Subplot pertama: gunakan garis putus-putus berwarna hitam.

    - Subplot kedua: gunakan garis solid (penuh) berwarna merah.

    - Subplot ketiga: gunakan penanda titik-titik (dots) berwarna biru.

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

dataRandom = np.random.randn(50,).cumsum()

ax1.plot(dataRandom, 'k--')
ax2.plot(dataRandom, 'r-')
ax3.plot(dataRandom, 'b.')
plt.show()
```
