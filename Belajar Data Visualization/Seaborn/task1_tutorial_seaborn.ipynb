2. Visualisasi Multivariat: Buat sebuah Scatter Plot menggunakan fungsi spesifik dari Seaborn, bukan Matplotlib.

    - Petakan beban_cpu_persen ke sumbu X dan suhu_server_celcius ke sumbu Y.

    - Syarat Mutlak (Core Mechanics): Bedakan warna setiap titik berdasarkan kolom kategorikal kategori_status. Kamu wajib menggunakan parameter bawaan Seaborn untuk melakukan pemetaan warna ini secara otomatis dalam satu baris kode, tanpa melakukan slicing DataFrame berulang kali seperti yang biasa dilakukan di Matplotlib.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

# Dummy Data: Simulasi metrik performa server
np.random.seed(42)
data = {
    'beban_cpu_persen': np.random.uniform(10, 100, 100),
    'suhu_server_celcius': np.random.uniform(20, 80, 100),
    'kategori_status': np.random.choice(['Aman', 'Peringatan', 'Kritis'], 100)
}
df = pd.DataFrame(data)
```

```python
# Perintah untuk mengaktifkan tema visual bawaan
sns.set_theme()

sns.scatterplot(data=df,
                x='beban_cpu_persen',
                y='suhu_server_celcius',
                hue='kategori_status')
plt.show()
```

## Tugas Praktik:
1. Gunakan dummy data df yang sama dari sesi sebelumnya.

2. Kalkulasi Korelasi: Buat sebuah variabel baru (misal: matriks_korelasi) yang berisi perhitungan korelasi antar fitur numerik di dalam df. (Petunjuk: Pandas memiliki fungsi bawaan .corr() untuk ini, pastikan kamu hanya memilih kolom bertipe numerik).

3. Visualisasi Heatmap: Panggil fungsi Seaborn untuk membuat Heatmap dari matriks tersebut.

4. Anotasi Data: Aktifkan parameter annot=True agar skor korelasinya (berupa angka desimal dari -1 hingga 1) tercetak tepat di tengah kotak warna.

5. Skema Warna: Gunakan skema warna yang divergen (misalnya cmap='coolwarm') agar korelasi negatif (biru) dan positif (merah) sangat kontras di mata.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

matriks_korelasi = df.select_dtypes(include='number').corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    matriks_korelasi,
    annot=True,
    cmap='coolwarm'
)

plt.show()
```
