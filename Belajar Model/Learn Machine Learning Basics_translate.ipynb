## **Building an End to End Machine Learning Project** | [👆](https://github.com/AdilShamim8/Machine_Learning_Housing_Corp)

#### 1. Bagaimana Anda mendefinisikan Machine Learning?
Machine Learning (Pembelajaran Mesin) adalah tentang membangun sistem yang dapat belajar dari data. Belajar di sini berarti menjadi lebih baik (meningkat kinerjanya) dalam melakukan suatu tugas, berdasarkan suatu tolok ukur kinerja tertentu.

#### 2. Bisakah Anda sebutkan empat jenis masalah di mana teknologi ini sangat unggul?
Machine Learning sangat unggul untuk menangani masalah kompleks yang tidak memiliki solusi algoritmik biasa, menggantikan daftar panjang aturan yang harus diatur secara manual (hand-tuned rules), membangun sistem yang dapat beradaptasi dengan lingkungan yang terus berubah-ubah, dan terakhir, membantu manusia dalam mempelajari sesuatu (misalnya, penggalian data atau data mining).

#### 3. Apa yang dimaksud dengan labeled training set (himpunan data pelatihan berlabel)?
Labeled training set adalah sekumpulan data pelatihan yang sudah mencakup solusi yang diinginkan (atau dikenal sebagai label) untuk setiap sampel/contoh data di dalamnya.

#### 4. Apa dua tugas supervised (pembelajaran terpandu) yang paling umum?
Dua tugas supervised yang paling umum adalah regresi (regression) dan klasifikasi (classification).

#### 5. Bisakah Anda sebutkan empat tugas unsupervised (pembelajaran tanpa panduan) yang umum?
Tugas-tugas unsupervised yang umum meliputi pengelompokan (clustering), visualisasi (visualization), reduksi dimensi (dimensionality reduction), dan pembelajaran aturan asosiasi (association rule learning).

#### 6. Jenis algoritma Machine Learning apa yang akan Anda gunakan untuk membuat robot bisa berjalan di berbagai medan yang tidak diketahui?
Reinforcement Learning (Pembelajaran Penguatan) kemungkinan besar akan memberikan hasil terbaik jika kita ingin robot belajar berjalan di berbagai medan yang tidak diketahui, karena ini adalah jenis masalah yang biasanya ditangani oleh Reinforcement Learning. Masalah ini mungkin saja dinyatakan sebagai masalah supervised atau semi-supervised learning, namun pendekatannya akan terasa kurang alami.

#### 7. Jenis algoritma apa yang akan Anda gunakan untuk mengelompokkan (mensegmentasikan) pelanggan Anda ke dalam beberapa grup?
Jika Anda tidak tahu cara menentukan grup-grup tersebut, maka Anda dapat menggunakan algoritma klasterisasi (clustering algorithm/unsupervised learning) untuk mensegmentasikan pelanggan Anda ke dalam kelompok-kelompok pelanggan yang serupa. Namun, jika Anda sudah tahu grup apa saja yang ingin Anda miliki, maka Anda dapat memberikan banyak contoh dari masing-masing grup tersebut ke algoritma klasifikasi (classification algorithm/supervised learning), dan algoritma tersebut akan mengklasifikasikan semua pelanggan Anda ke dalam grup-grup ini.

#### 8. Apakah Anda akan membingkai masalah deteksi spam sebagai masalah supervised learning atau unsupervised learning?
Deteksi spam adalah contoh masalah supervised learning (pembelajaran terpandu) yang khas: algoritma diberikan banyak email beserta labelnya (spam atau bukan spam).

#### 9. Apa yang dimaksud dengan online learning system (sistem pembelajaran daring/inkremental)?
Online learning system adalah sistem yang dapat belajar secara bertahap (inkremental), berbeda dengan batch learning system (sistem pembelajaran massal). Hal ini membuatnya mampu beradaptasi dengan cepat terhadap perubahan data maupun sistem otonom, serta dapat digunakan untuk melatih data dalam jumlah yang sangat besar.

#### 10. Apa yang dimaksud dengan out-of-core learning?
Algoritma out-of-core dapat menangani data dalam jumlah sangat besar yang tidak muat di dalam memori utama (RAM) komputer. Algoritma out-of-core learning memotong data tersebut menjadi beberapa bagian kecil (mini-batches) dan menggunakan teknik online learning untuk belajar dari bagian-bagian kecil tersebut.

#### 11. Jenis algoritma pembelajaran apa yang mengandalkan ukuran kemiripan (similarity measure) untuk membuat prediksi?
Sistem instance-based learning (pembelajaran berbasis contoh) mempelajari data pelatihan dengan cara menghafalnya; kemudian, ketika diberikan sampel data baru, sistem ini menggunakan ukuran kemiripan untuk menemukan sampel data yang paling mirip yang telah dihafal sebelumnya, lalu menggunakannya untuk membuat prediksi.

#### 12. Apa perbedaan antara model parameter (parameter model) dan hyperparameter dari algoritma pembelajaran?
Sebuah model memiliki satu atau lebih parameter model yang menentukan apa yang akan diprediksinya ketika diberikan sampel data baru (misalnya, kemiringan atau slope pada model linear). Algoritma pembelajaran berusaha menemukan nilai optimal untuk parameter-parameter ini agar model dapat menggeneralisasi dengan baik pada sampel data baru. Sementara itu, hyperparameter adalah parameter dari algoritma pembelajarannya itu sendiri, bukan dari modelnya (misalnya, tingkat regularisasi yang akan diterapkan).

#### 13. Apa yang dicari oleh algoritma model-based learning? Apa strategi paling umum yang mereka gunakan agar berhasil? Bagaimana cara mereka membuat prediksi?
Algoritma model-based learning (pembelajaran berbasis model) mencari nilai optimal untuk parameter model agar model tersebut dapat menggeneralisasi dengan baik pada sampel data baru. Kita biasanya melatih sistem seperti ini dengan meminimalkan fungsi biaya (cost function) yang mengukur seberapa buruk kinerja sistem dalam membuat prediksi pada data pelatihan, ditambah dengan penalti untuk kompleksitas model jika model tersebut diregularisasi. Untuk membuat prediksi, kita memasukkan fitur-fitur dari sampel data baru ke dalam fungsi prediksi model, dengan menggunakan nilai-nilai parameter yang telah ditemukan oleh algoritma pembelajaran tersebut.

#### 14. Bisakah Anda sebutkan empat tantangan utama dalam Machine Learning?
Beberapa tantangan utama dalam Machine Learning adalah kurangnya data, kualitas data yang buruk, data yang tidak representatif, fitur-fitur yang tidak informatif, model yang terlalu sederhana sehingga mengalami underfitting pada data pelatihan, serta model yang terlalu kompleks sehingga mengalami overfitting pada data.

#### 15. Jika model Anda berkinerja sangat baik pada data pelatihan tetapi memiliki generalisasi yang buruk pada sampel data baru, apa yang sebenarnya terjadi? Bisakah Anda sebutkan tiga solusi yang memungkinkan?
Jika sebuah model berkinerja sangat baik pada data pelatihan tetapi memiliki generalisasi yang buruk pada sampel data baru, model tersebut kemungkinan besar mengalami overfitting terhadap data pelatihan (atau kita hanya sedang sangat beruntung dengan data pelatihan tersebut). Solusi yang memungkinkan untuk mengatasi overfitting adalah mengumpulkan lebih banyak data, menyederhanakan model (memilih algoritma yang lebih sederhana, mengurangi jumlah parameter atau fitur yang digunakan, atau meregularisasi model), atau mengurangi derau (noise) pada data pelatihan.

#### 16. Apa yang dimaksud dengan test set (himpunan data uji), dan mengapa Anda ingin menggunakannya?
Test set digunakan untuk mengestimasi kesalahan generalisasi (generalization error) yang akan dihasilkan oleh model pada sampel data baru, sebelum model tersebut diluncurkan ke tahap produksi.

#### 17. Apa tujuan dari validation set (himpunan data validasi)?
Validation set digunakan untuk membandingkan berbagai model. Keberadaannya memungkinkan kita untuk memilih model terbaik dan mengatur (menala) hyperparameter-nya.

#### Apa yang dimaksud dengan train-dev set, kapan Anda membutuhkannya, dan bagaimana cara menggunakannya?
Train-dev set digunakan ketika ada risiko ketidaksesuaian (mismatch) antara data pelatihan (training data) dan data yang digunakan dalam dataset validasi serta pengujian (yang seharusnya selalu semirip mungkin dengan data yang akan dihadapi saat model sudah berada di tahap produksi). Train-dev set adalah bagian dari training set yang disisihkan (model tidak dilatih menggunakan bagian ini). Model dilatih pada sisa training set yang ada, lalu dievaluasi pada train-dev set sekaligus validation set. Jika model berkinerja baik pada training set tetapi buruk pada train-dev set, maka model tersebut kemungkinan besar mengalami overfitting terhadap training set. Jika model berkinerja baik pada training set sekaligus train-dev set, tetapi buruk pada validation set, maka kemungkinan besar terdapat ketidaksesuaian data (data mismatch) yang signifikan antara data pelatihan dan data validasi + uji. Jika hal itu terjadi, Anda harus mencoba memperbaiki data pelatihan agar terlihat lebih mirip dengan data validasi + uji.

Catatan Tambahan untuk Membantu Analisis Data Mismatch vs Overfitting:

Untuk mempermudah pemahaman kapan masalah terjadi, Anda bisa merujuk pada panduan evaluasi performa berikut:

- Kasus A (Overfitting):

    - Training Set Error: Rendah

    - Train-Dev Set Error: Tinggi

    - Validation Set Error: Tinggi
    (Kesimpulan: Model menghafal data pelatihan tetapi gagal menggeneralisasi ke data baru dari distribusi yang sama).

- Kasus B (Data Mismatch):

    - Training Set Error: Rendah

    - Train-Dev Set Error: Rendah

    - Validation Set Error: Tinggi
    (Kesimpulan: Model berhasil menggeneralisasi ke data baru dari distribusi yang sama, tetapi gagal karena distribusi data validasi/produksi berbeda dengan data pelatihan).

#### 19. Apa yang bisa salah jika Anda mengatur (menala) hyperparameter menggunakan test set?
Jika Anda menala hyperparameter menggunakan test set, Anda berisiko mengalami overfitting terhadap test set tersebut, dan kesalahan generalisasi (generalization error) yang Anda ukur akan menjadi terlalu optimis (Anda mungkin akan meluncurkan model yang kinerjanya ternyata lebih buruk daripada yang Anda harapkan).

Penjelasan Singkat:

Proses ini sering disebut sebagai data leakage (kebocoran data) pada tahap evaluasi. Ketika kita berulang kali mengubah hyperparameter berdasarkan hasil test set, informasi dari test set secara tidak sengaja "bocor" ke dalam proses pemilihan model. Akibatnya, test set tersebut tidak lagi menjadi refleksi yang jujur dan independen untuk menguji data masa depan (unseen data). Itulah mengapa fungsi ini harus didelegasikan sepenuhnya kepada validation set.
