## **1. Diferensiasi tuh Apaan Sih?**

**Konsepnya gini:** Diferensiasi itu buat ngukur *laju perubahan* sebuah fungsi, nih.

Jika  
\[
$y = f(x)$,
\]  
Nah, turunan $f'(x)$ itu bakal ngasih tau lu seberapa cepet si $y$ berubah pas lu geser $x$ dikit aja.

**Kalo liat gambar:** itu tuh kemiringan garis singgung di kurvanya, oke?

> **Cek ombak dulu:**  
> Misal $f(x)=x^2$, berapa coba kemiringannya pas $x=2$?  
> \[
> $f'(x)=2x \Rightarrow f'(2)=4$.
> \]

---

## **2. Turunannya Konstanta**

\[
$\frac{d}{dx}(c) = 0$
\]

Konstanta mah diem aja gak berubah -> kemiringannya ya 0.

---

## **3. Aturan Pangkat (Power Rule)**

Jika  
\[
$f(x)=x^n$,
\]

maka  
\[
$\frac{d}{dx}(x^n)=n x^{n-1}$.
\]

Contoh:  
\[
$\frac{d}{dx}(x^3)=3x^2$.
\]

---

## **4. Aturan Tambah-tambahan**

\[
$\frac{d}{dx}[f(x)+g(x)] = f'(x) + g'(x)$
\]

---

## **5. Aturan Kali-kalian (Product Rule)**

\[
$\frac{d}{dx}[f(x)g(x)] = f'(x)g(x) + f(x)g'(x)$
\]

---

## **6. Aturan Bagi-bagian (Quotient Rule)**

\[
$\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right]
= \frac{f'(x)g(x) - f(x)g'(x)}{(g(x))^2}$
\]

---

## **7. Aturan Rantai (Chain Rule)**

Untuk komposisi \( f(g(x)) \):

\[
$\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)$
\]

**Hubungannya ama ML:** ini nih dasarnya *backpropagation*, cuy.

---

## **8. Turunan Parsial (Potong-potong)**

Buat fungsi yang variabelnya rame-rame $f(x,y)$:

- $\frac{\partial f}{\partial x}$: turunin buat $x$ aja, anggep si $y$ itu patung (konstan).  
- $\frac{\partial f}{\partial y}$: turunin buat $y$ aja, anggep si $x$ itu patung (konstan).

Contoh:  
\[
$f(x,y)=x^2 y + y^3$
\]

\[
$\frac{\partial f}{\partial x}=2xy$
\]

\[
$\frac{\partial f}{\partial y}=x^2 + 3y^2$
\]

---

## **9. Turunan Tingkat Tinggi (Lanjut Terus)**

Turunan kedua:  
\[
$f''(x)$
\]  
ngasih tau kita soal seberapa melengkung kurvanya.

**Koneksi ML:** Hessian dipakai di optimisasi lanjutan.

---

## **10. Diferensiasi Matriks (Pro Banget)**

- Jika  
  \[
  $f(x)=a^\top x$,
  \]  
  maka  
  \[
  $\nabla_x f = a$.
  \]

- Jika  
  \[
  $f(x)=x^\top A x$,
  \]  
  maka  
  \[
  $\nabla_x f = (A + A^\top)x$.
  \]

**Koneksi ML:** gradien loss function, training neural network, optimisasi.

> # `Teori Optimasi (Biar Makin Jago)`

# Fungsi & Pengoptimalan — Catatan

---

## 1. Fungsi

Sebuah **fungsi** memetakan masukan ke keluaran:

$$
y = f(x)
$$

Dalam ML, $f$ sering kali menunjukkan model (model linier, neural network, dll.).

**Contoh (regresi linier):**

$$
y = Xw + b \quad\text{atau}\quad \hat{y} = w^\top x + b
$$

---

## 2. Fungsi Multivariasi

Fungsi banyak variabel:

$$
f(x_1, x_2, \dots, x_n)
$$

---

## 3. Parameter Suatu Fungsi

Contoh:

$$
f(x) = w_1 x_1 + w_2 x_2 + b
$$

---

## 4. Maksimum & Minimum

Kondisi order pertama:

$$
\nabla f(x) = 0
$$

---

## 5. Fungsi Kerugian

- MSE:

$$
L = \frac{1}{n} \sum_i (y_i - \hat{y}_i)^2
$$

- Cross-entropy:

$$
L = -\sum_i y_i \log \hat{p}_i
$$

---

## 6. Memilih Loss Function

**Kenapa cross-entropy lebih bagus untuk klasifikasi?**

- Selaras dengan distribusi kategorikal + softmax  
- Gradien lebih kuat saat salah prediksi  
- Konvergensi lebih stabil dari MSE  

---

## 7. Menghitung Parameter

Mencari $w$ yang meminimalkan $L(w)$ menggunakan gradient-based optimization.

---

## 8. Kerugian Cembung & Cekung

Cembung → minimum global tunggal.

---

## 9. Penurunan Gradien

$$
w \leftarrow w - \eta \,\nabla_w L(w)
$$

---

## 10. Hessian (Percepatan)

$$
H_{ij} = \frac{\partial^2 f}{\partial x_i \partial x_j}
$$

---

## 11. Masalah Optimasi

- Local minima  
- Saddle points  
- Vanishing/exploding gradients  

---

## 12. Optimasi Terbatas

Contoh: Lasso

$$
\|w\|_1 \le \alpha
$$
