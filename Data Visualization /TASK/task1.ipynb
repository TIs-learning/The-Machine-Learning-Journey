## Fase 1: Fondasi Teori & Statistik Dasar (Level 0)
Sebelum menyentuh kode, kamu harus paham mengapa dan kapan sebuah grafik digunakan.

- Pahami Tipe Data: Bedakan antara data kategorikal (nominal, ordinal) dan numerik (diskrit, kontinu).

- Pilih Grafik yang Tepat: Pelajari kapan harus menggunakan Bar chart (perbandingan), Line chart (tren waktu), Scatter plot (hubungan dua variabel), dan Histogram (distribusi data).

- Statistik Deskriptif: Pahami konsep mean, median, variance, standard deviation, dan kuartil (Q1, Q2, Q3).

### Peta Ekosistem Visualisasi Python

Setiap *library* diciptakan dengan filosofi dan target penyelesaian masalah yang sama sekali berbeda. Menggunakan alat yang salah untuk tugas tertentu akan menghancurkan efisiensi kodemu.

**1. Matplotlib: Sang Fondasi (Low-Level)**
Ini adalah "bahasa mesin"-nya visualisasi Python. Matplotlib memberimu kontrol absolut atas setiap piksel, garis, margin, dan kanvas.

* **Kekuatan:** Sangat kustomisasi, sangat stabil, dan merupakan fondasi tulang punggung dari banyak *library* lain di Python.
* **Kelemahan:** Kodenya panjang dan kaku (*verbose*). Untuk membuat grafik yang terlihat modern, kamu butuh puluhan baris kode hanya untuk modifikasi estetika.
* **Penggunaan Ideal:** Saat kamu butuh arsitektur grafik yang sangat spesifik, kustomisasi tingkat piksel, dan tidak butuh interaksi.

**2. Seaborn: Sang Detektif Statistik (High-Level)**
Seaborn dibangun tepat di atas Matplotlib. Filosofinya adalah mereduksi puluhan baris kode kotor Matplotlib menjadi satu atau dua baris kode elegan yang khusus ditujukan untuk kebutuhan analisis statistik.

* **Kekuatan:** Sangat rapi secara bawaan (*default*). Memiliki fungsi bawaan untuk kebutuhan *Machine Learning* fase awal, seperti deteksi korelasi (*heatmap*), pemetaan distribusi (*violin plot*), dan *pairplot*.
* **Kelemahan:** Kontrol tingkat rendahnya terbatas. Jika kamu butuh modifikasi ekstrem pada tata letaknya, kamu tetap harus membongkar dan memanggil objek Matplotlib di baliknya.
* **Penggunaan Ideal:** *Exploratory Data Analysis* (EDA) untuk menemukan pola matematis tersembunyi sebelum melatih algoritma model.

**3. Plotly: Sang Interaktif (Web-Based)**
Plotly tidak berakar pada Matplotlib; ia menggunakan mesin *rendering* JavaScript (Plotly.js) di belakang layar meskipun kamu menulisnya dengan Python.

* **Kekuatan:** Interaktivitas penuh secara otomatis. Pengguna bisa melakukan *zoom*, *hover* untuk melihat detail metrik, dan *panning* langsung di atas grafik tanpa perlu deklarasi mode interaktif khusus (seperti `%matplotlib widget`).
* **Kelemahan:** Mengkonsumsi memori yang jauh lebih besar. Hasil akhirnya berupa file HTML atau dirender di *browser*, sehingga komputasinya lebih berat jika data yang dimasukkan terlalu masif.
* **Penggunaan Ideal:** Pembuatan *dashboard* presentasi teknis di mana pemangku kepentingan perlu berinteraksi langsung dengan grafik modelmu.

**4. Bokeh: Sang Real-Time (Web-Based)**
Sama seperti Plotly, Bokeh menargetkan *browser* web. Namun, arsitektur *backend* Bokeh didesain lebih agresif untuk menangani aliran data (*streaming*) dalam volume besar.

* **Kekuatan:** Performa sangat tangguh untuk dataset masif. Mampu memperbarui titik grafik secara *live* langsung dari komunikasi *server*.
* **Kelemahan:** Sintaks dan paradigmanya memiliki kurva belajar yang cukup curam, serta ekosistem komunitasnya tidak sebesar Plotly.
* **Penggunaan Ideal:** Dasbor analitik *live* atau aplikasi web untuk memonitor fluktuasi metrik *real-time*.
