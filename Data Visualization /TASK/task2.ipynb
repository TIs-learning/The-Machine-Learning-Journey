## Fase 2: Penguasaan Library Dasar Python (Beginner)
Python adalah bahasa standar untuk ML. Di fase ini, kamu mulai mengubah data menjadi visual.

- Pandas: Pelajari cara memanipulasi data (dataframe) dan menggunakan fungsi plot bawaan Pandas yang cepat.

- Matplotlib: Ini adalah fondasi visualisasi di Python. Pelajari cara membuat figure, axes, mengatur label, judul, dan warna.

- Seaborn: Library yang dibangun di atas Matplotlib khusus untuk visualisasi statistik. Sangat direkomendasikan untuk ML pemula karena kodenya lebih ringkas dan tampilannya lebih rapi.

### 1. Menginvestigasi Relasi (Hubungan Antar-Variabel)

Grafik di kategori ini digunakan saat kamu ingin membuktikan apakah variabel A memengaruhi variabel B. Ini adalah langkah wajib dalam rekayasa fitur (*feature engineering*) untuk mencari bobot matematis yang kuat sebelum melatih model.

* **Scatter Plot (Grafik Pencar):** Memetakan perpotongan dua variabel numerik (Sumbu X dan Y).
* *Fungsi ML:* Sangat krusial untuk mendeteksi apakah hubungan datamu linear (bisa diselesaikan dengan Regresi Linear) atau membentuk kelompok spasial (membutuhkan algoritma *Clustering* seperti K-Means).


* **Heatmap (Peta Panas):** Matriks warna yang menunjukkan skor korelasi antar semua fitur sekaligus.
* *Fungsi ML:* Digunakan untuk mendeteksi multikolinearitas. Jika ada dua fitur yang warnanya terlalu identik (korelasinya nyaris 1), salah satunya harus dibuang dari arsitektur datamu agar kalkulasi algoritma tidak redundan.



### 2. Membedah Distribusi (Sebaran Data)

Sebelum menyusun *pipeline* pemrosesan data, kamu harus tahu bentuk asli datamu. Apakah normal, miring (*skewed*), atau penuh anomali?

* **Histogram:** Memotong rentang data numerik menjadi beberapa "keranjang" (*bins*) dan menghitung frekuensi di setiap keranjangnya.
* *Fungsi ML:* Mengecek apakah datamu berdistribusi normal. Banyak algoritma fundamental berasumsi datamu berdistribusi normal; jika bentuk histogrammu sangat miring ke kiri/kanan, kamu wajib melakukan transformasi logaritmik terlebih dahulu.


* **Box Plot (Diagram Kotak Garis):** Representasi visual dari statistik kuartil (Q1, Q2/Median, Q3).
* *Fungsi ML:* Ini adalah alat *debugging* utama untuk mendeteksi *outliers* (pencilan). Titik-titik yang jatuh di luar "kumis" (*whiskers*) dari Box Plot adalah anomali yang berpotensi merusak akurasi prediksi modelmu jika tidak ditangani secara khusus.



### 3. Menganalisis Tren Kontinu (Waktu / Iterasi)

Kategori ini khusus untuk metrik yang saling terhubung secara berurutan dan tidak bisa diacak.

* **Line Chart (Diagram Garis):** Titik data yang saling dihubungkan dengan garis solid.
* *Fungsi ML:* Jangan pernah menggunakan ini untuk data kategori acak. Gunakan murni untuk data runtun waktu (*Time-Series* seperti harga saham) atau untuk membuat kurva *Loss/Accuracy* dalam mengevaluasi metrik performa model pada setiap *epoch* (iterasi pembelajaran).



### 4. Membandingkan Kategori (Komparasi & Komposisi)

Kategori ini membandingkan besaran nilai antar kelompok diskrit (tidak kontinu).

* **Bar Chart (Diagram Batang):** Panjang batang mewakili kuantitas.
* *Fungsi ML:* Sempurna untuk membandingkan metrik evaluasi akhir antar arsitektur. Misalnya, menjejerkan akurasi algoritma *Random Forest*, *Support Vector Machine*, dan *Neural Network* pada dataset yang sama.


* **Pie Chart (Diagram Lingkaran):** Membagi lingkaran berdasarkan persentase.
* *Kritik Fundamental:* Hindari menggunakan grafik ini di lingkungan analitik tingkat lanjut. Mata manusia secara biomekanis sangat buruk dalam membedakan luasan sudut yang berdekatan (misalnya membedakan antara 33% dan 36% dalam bentuk potongan *pie*). Jika kamu ingin membandingkan komposisi, *Bar Chart* selalu lebih presisi dan mudah dicerna logika.
