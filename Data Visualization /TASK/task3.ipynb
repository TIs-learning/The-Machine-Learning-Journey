## Fase 3: Exploratory Data Analysis / EDA (Intermediate)
Di sini kamu menggunakan visualisasi sebagai "detektif" untuk mempersiapkan data sebelum dimasukkan ke model ML.

- Identifikasi Outliers (Pencilan): Gunakan Box plot atau Violin plot untuk melihat data ekstrem yang bisa merusak modelmu.

- Analisis Korelasi: Gunakan Heatmap (dari Seaborn) untuk melihat fitur mana yang memiliki korelasi kuat dengan variabel target, atau untuk mendeteksi multikolinearitas antar fitur.

- Distribusi Variabel: Gunakan KDE plot (Kernel Density Estimation) atau Histogram untuk melihat apakah datamu terdistribusi normal atau miring (skewed).
