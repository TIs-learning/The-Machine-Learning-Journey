## Tugas Instruksional Anatomi Matplotlib:
Tuliskan kelanjutan kode di atas dengan aturan ketat berikut. Semuanya wajib menggunakan pendekatan Object-Oriented (melalui objek ax atau fig), dilarang menggunakan perintah State-based (seperti plt.plot atau plt.title):

1. Inisiasi: Buat kanvas utama dan axes (fig, ax).

2. Plotting Dasar: Plot data dari kolom Close. Beri argumen warna hijau (color='green') dan berikan nama identitas pada garis tersebut (label='Harga Penutupan').

3. Titik Anatomi 1 (Judul): Tambahkan judul grafik "Pergerakan Harga Saham FB". (Petunjuk: Di pendekatan OO, nama fungsinya bukan title()).

4. Titik Anatomi 2 (Label Sumbu): Tambahkan teks "Tanggal" untuk sumbu X dan "Harga (USD)" untuk sumbu Y. (Petunjuk: Gunakan metode setter dari objek ax).

5. Aktivasi Komponen: Munculkan Legend (kotak keterangan yang berisi tulisan "Harga Penutupan" dari langkah 2) ke dalam kanvas.

6. Finishing: Panggil fungsi autofmt_xdate() untuk merapikan tanggal, lalu tampilkan kanvasnya ke layar.

```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt

# --- FASE 1: PREPARASI DATA (SUDAH DISEDIAKAN) ---
# Membaca file yang baru saja kamu unggah
df = pd.read_csv('FBCopy.csv')

# Mengonversi string menjadi format waktu dan menjadikannya indeks
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date').sort_index()

# --- FASE 2: VISUALISASI MATPLOTLIB (TUGASMU) ---
# Lanjutkan kode di bawah ini...
closePrice = df['Close']

# inisiasi awal kanvas
fig, ax= plt.subplots()

# membuat kotak grafik
ax.set_ylabel('Harga (USD)')
ax.set_xlabel('Tanggal')
ax.set_title('Pergerakan Harga Saham FB')
ax.plot(closePrice, color='green', linestyle='-', label='Harga Penutupan')
ax.legend()
fig.autofmt_xdate()
fig.tight_layout()

# Terakhir
plt.show()
```
