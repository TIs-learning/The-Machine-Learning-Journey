## Fase 4: Visualisasi Dimensi Tinggi & Interaktif (Advanced)
Data ML seringkali memiliki puluhan bahkan ratusan fitur (dimensi), yang tidak bisa digambar dengan grafik 2D atau 3D biasa.

- Reduksi Dimensi (PCA, t-SNE, UMAP): Pelajari cara memvisualisasikan data berdimensi tinggi ke dalam bentuk 2D atau 3D untuk melihat pengelompokan (clustering) secara alami.

- Visualisasi Interaktif: Pelajari Plotly atau Bokeh. Grafik interaktif memungkinkan kamu melakukan zoom, hover untuk melihat detail data, yang sangat berguna saat melakukan presentasi teknis.

- Pembuatan Dashboard: Kenali Streamlit atau Dash untuk mengubah visualisasi ML kamu menjadi aplikasi web sederhana yang bisa dicoba orang lain.
