## Fase 5: Visualisasi Evaluasi & Interpretasi Model ML (Expert)
Fase ini berfokus pada visualisasi performa model (apakah modelmu bagus?) dan Explainable AI (mengapa modelmu memutuskan hal tersebut?).

- Evaluasi Klasifikasi: Pelajari cara membaca Confusion Matrix (untuk melihat salah tebak dari model), ROC-AUC Curve, dan Precision-Recall Curve.

- Evaluasi Regresi: Gunakan Residual Plot (membandingkan error prediksi) dan Actual vs Predicted plot.

- Explainable AI (XAI): Pelajari SHAP (SHapley Additive exPlanations) dan LIME. Visualisasi ini sangat advance dan digunakan untuk menjelaskan dampak setiap fitur terhadap keputusan akhir model secara individual (misal: mengapa model menolak pinjaman nasabah A).
