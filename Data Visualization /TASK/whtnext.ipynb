Jika kamu bertanya apa yang **belum** kamu pelajari, daftarnya masih panjang dan ini berada di ranah *Advanced Visualization*. Berikut adalah celah yang masih kosong dalam kerangka belajarmu:

### 1. Visualisasi Ruang Dimensi Tinggi (Dimensionality Reduction)

Sejauh ini, kamu baru memetakan dua atau tiga variabel ke dalam sumbu X, Y, dan warna (seperti *Scatter plot* CPU vs Suhu). Di dunia nyata, dataset bisa memiliki 50 hingga 100 fitur. Kamu tidak bisa membuat grafik 100 dimensi.

* **Yang belum kamu pelajari:** Bagaimana memampatkan 100 dimensi tersebut ke dalam kanvas 2D tanpa menghilangkan informasi matematisnya menggunakan algoritma **PCA (Principal Component Analysis)**, **t-SNE**, atau **UMAP**. Ini adalah visualisasi tingkat lanjut untuk melihat apakah algoritma klasifikasimu sebenarnya punya peluang untuk berhasil atau tidak.

### 2. Explainable AI (XAI) & Visualisasi Interpretasi Model

Jika model prediksi kegagalan servermu tiba-tiba memvonis sebuah server akan meledak besok, teknisi di lapangan akan bertanya: *"Kenapa modelmu menyimpulkan begitu?"*
Kamu tidak bisa menjawabnya dengan grafik Seaborn biasa.

* **Yang belum kamu pelajari:** Kamu belum menyentuh instrumen seperti **SHAP (SHapley Additive exPlanations)** atau **LIME**. Visualisasi ini membedah *black-box* algoritma (terutama algoritma kompleks seperti *ensemble methods* atau *random forest*) untuk menunjukkan secara persis fitur mana yang mendorong prediksi ke atas, dan fitur mana yang menahannya ke bawah untuk setiap satu baris prediksi.

### 3. Visualisasi Evaluasi Batas Keputusan (Decision Boundaries)

Saat kamu memanggil fungsi dari *library* ML (seperti Scikit-learn) dan melatih model, model itu akan membuat garis batas pemisah antar-kelas di dalam otaknya.

* **Yang belum kamu pelajari:** Memvisualisasikan *Decision Boundary*. Ini adalah teknik menggambar kontur area prediksi model di atas *scatter plot* datamu. Dengan melihat visualisasi kontur ini, kamu bisa mendiagnosis secara langsung apakah modelmu menderita *Overfitting* (batasnya terlalu berkelok-kelok menghafal data) atau *Underfitting* (batasnya berupa garis lurus kaku yang mengabaikan pola).

### 4. Kurva Diagnostik Performa Algoritma

Melihat skor akurasi berupa angka $90\%$ tidak cukup. Kamu harus bisa memvisualisasikan kelemahan modelmu.

* **Yang belum kamu pelajari:** Menggambar matriks dan kurva evaluasi secara manual dari hasil prediksi model. Ini termasuk *ROC-AUC Curve* (untuk melihat keseimbangan antara *True Positive* dan *False Positive*) serta *Precision-Recall Curve* (sangat krusial untuk dataset di mana kelas targetnya sangat tidak seimbang).

---

Meninggalkan visualisasi dasar sekarang memang keputusan yang tepat agar kita bisa mulai masuk ke arsitektur pemodelan. Namun, ingatlah bahwa kita akan memanggil kembali library visualisasi ini sebagai alat "bedah" saat model kita nanti gagal memberikan prediksi yang masuk akal.

Dari keempat konsep visualisasi tingkat lanjut di atas (Reduksi Dimensi, XAI, *Decision Boundary*, atau Kurva Diagnostik), mana yang menurut logikamu paling menantang dan ingin kamu jadikan jembatan pertama kita saat masuk ke materi pemodelan nanti?
