## Categorical Handling
(Label Encoding & One Hot Encoding)

1. Categorical Encoding: proses mengkonversi niali kategorical menjadi nilai numerical 
2. jenisnya ada banyak tapi populer ada 2, yaitu: Label encoding dan One hot encoding (referensi: https:// -en.wikipedia.org/wiki/one-hot)
3. Di label encoding setiap kategori feature akan diurutkan secara alfabet dan digambarkan dengan nilai integer
4. di one hot encoding kategori feature akan diurutkan secraa alfabet dan digambarkan sbgai sekumpulan bits


## Label Encoding

```python
# Buat dataet
import pandas as pd

df = pd.DataFrame({
    'country':['India', 'US', 'Japan', 'US', 'Japan'],
    'age':[44, 34, 46, 35, 23],
    'salary':[72000, 65000, 98000, 45000, 34000]
})
df
```

## Label Encoding pada Scikit Learn

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['country'] = label_encoder.fit_transform(df['country'])
df
```

```python
label_encoder.classes_
```

## One Hot Encoding

```python
# Buat dataet
import pandas as pd

df = pd.DataFrame({
    'country':['Indonesia', 'UK', 'Jepang', 'UK', 'Jepang'],
    'age':[44, 34, 46, 35, 23],
    'salary':[72000, 65000, 98000, 45000, 34000]
})
df
```

## One Hot encoding pada scikit learn

```python
from sklearn.preprocessing import OneHotEncoder
x = df['country'].values.reshape(-1, 1)
x
```

```python
oneshot_encoder = OneHotEncoder()
x = oneshot_encoder.fit_transform(x).toarray()
x
```

```python
oneshot_encoder.categories_
```

```python
df_onehot = pd.DataFrame(x, columns=[str(i) for i in range(x.shape[1])])
df_onehot
```

```python
df = pd.concat([df_onehot, df], axis=1)
df
```

```python
df = df.drop(['country'], axis=1)
df
```

## Notes: Label Encoding vs One Hot Encoding
Kita menerapkan One Hot Encoding bila:

- Nilai categorical adalah nominal
- Jumlah kategori yang ada tidak terlalu banyak

Kita menerapkan Label Encoding bila:

- Nilai categorical adalah ordinal
- Jumlah kategori yang ada relatif banyak
