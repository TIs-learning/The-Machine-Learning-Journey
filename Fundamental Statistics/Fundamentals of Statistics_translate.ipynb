> # `Jenis-Jenis Utama Statistika`

### 1. **Statistika Deskriptif**

* **Tujuan:** Untuk **meringkas dan mendeskripsikan** data.
* **Apa yang dilakukan:** Menjelaskan *apa yang ditunjukkan oleh data* (tanpa membuat prediksi atau generalisasi).
* **Contoh:**

  * Mean (rata-rata), median, modus (ukuran pemusatan data).
  * Varians, standar deviasi, rentang (range), IQR (ukuran penyebaran data).
  * Grafik & diagram: histogram, diagram batang, boxplot.
* **Penggunaan dalam ML:** Memahami distribusi dataset, mendeteksi outlier, dan melakukan penskalaan fitur (feature scaling).

---

### 2. **Statistika Inferensial**

* **Tujuan:** Untuk **membuat prediksi, mengambil keputusan, atau melakukan generalisasi** terhadap suatu populasi berdasarkan sampel.
* **Apa yang dilakukan:** Menggunakan teori probabilitas untuk **menginferensikan** karakteristik kelompok yang lebih besar.
* **Contoh:**

  * Pengujian hipotesis (uji-t, chi-square, ANOVA).
  * Interval kepercayaan (confidence interval).
  * Analisis regresi.
* **Penggunaan dalam ML:**

  * Membandingkan performa dua model (apakah model A lebih baik daripada B?).
  * Menentukan apakah suatu fitur memiliki daya prediksi yang signifikan.
  * Menggeneralisasikan hasil dari data pelatihan ke data uji yang belum pernah dilihat sebelumnya.

---

### Subbidang yang Sering Disebutkan

* **Statistika Prediktif (Predictive Statistics)** → berfokus pada pembuatan prediksi (erat kaitannya dengan machine learning).
* **Analisis Data Eksploratif (Exploratory Data Analysis / EDA)** → bagian dari statistika deskriptif yang digunakan untuk **menemukan pola** dalam data.
* **Statistika Bayesian** → alternatif dari inferensi klasik (frequentist); menggunakan pengetahuan awal (prior) dan data untuk memperbarui keyakinan atau estimasi.

---

**Analogi Singkat:**

* *Statistika deskriptif* = "Kelas yang berisi 30 siswa ini memiliki tinggi rata-rata 165 cm."
* *Statistika inferensial* = "Berdasarkan sampel ini, kita memperkirakan bahwa tinggi rata-rata **seluruh siswa di sekolah** adalah sekitar 165 cm, dengan interval kepercayaan 95%."

Data secara umum dapat dibagi menjadi **Data Kualitatif (Kategorikal)** dan **Data Kuantitatif (Numerik)**.

---

### 1. **Data Kualitatif (Kategorikal)**

* Merepresentasikan **karakteristik, kategori, atau label** (bukan angka yang dapat dihitung secara matematis).
* **Subjenis:**

  * **Data Nominal** → kategori **tanpa urutan**.

    * Contoh: jenis kelamin (laki-laki/perempuan), warna (merah/biru/hijau), nama kota.
    * Penggunaan dalam ML: one-hot encoding sebelum dimasukkan ke model.
  * **Data Ordinal** → kategori **yang memiliki urutan atau tingkatan alami**.

    * Contoh: penilaian (buruk/cukup/baik/sangat baik), tingkat pendidikan (SMA, sarjana, magister).
    * Penggunaan dalam ML: terkadang label encoding dapat digunakan karena urutan kategorinya bermakna.

---

### 2. **Data Kuantitatif (Numerik)**

* Merepresentasikan **angka yang dapat diukur atau dihitung**.
* **Subjenis:**

  * **Data Diskret** → nilai yang dapat dihitung (bilangan bulat).

    * Contoh: jumlah siswa dalam kelas, jumlah mobil di tempat parkir.
    * Penggunaan dalam ML: model Poisson, fitur untuk klasifikasi.
  * **Data Kontinu** → nilai yang dapat diukur dan dapat mengambil **nilai apa pun dalam suatu rentang**.

    * Contoh: tinggi badan, berat badan, suhu, waktu.
    * Penggunaan dalam ML: model regresi, normalisasi, dan penskalaan data.

---

| Kuantitatif | Diskret | Jumlah siswa, jumlah gol yang dicetak | Model data hitungan |
|            | Kontinu | Tinggi badan, berat badan, suhu | Regresi, penskalaan |

---

**Analogi Singkat:**

* *Nominal* = Hanya berupa nama atau kategori (tanpa urutan).
* *Ordinal* = Urutan atau tingkatan memiliki makna.
* *Diskret* = Angka hasil menghitung.
* *Kontinu* = Angka hasil pengukuran.

> # `Ukuran Pemusatan Data (Measure of Central Tendency)`

<details>
<summary>Klik untuk melihat</summary>
    
### Definisi

* **Ukuran pemusatan data (central tendency)** adalah suatu nilai tunggal yang merepresentasikan **pusat** atau **nilai khas** dari suatu dataset.
* Ukuran ini membantu kita memahami **di mana sebagian besar titik data berada**.

---

### Ukuran Utama

### 1. **Mean (Rata-Rata Aritmetika)**

* Rumus:

  $$
  \bar{x} = \frac{\sum_{i=1}^n x_i}{n}
  $$
* Contoh: nilai = \[10, 20, 30] → mean = (10+20+30)/3 = 20
* **Penggunaan dalam ML:** penskalaan fitur (feature scaling), menghitung rata-rata kesalahan (MSE).
* **Keterbatasan:** sensitif terhadap outlier (misalnya pada data pendapatan).

---

### 2. **Median (Nilai Tengah)**

* Urutkan data, lalu ambil nilai yang berada di tengah.

  * Jika $n$ ganjil: ambil nilai tengah.
  * Jika $n$ genap: ambil rata-rata dari dua nilai tengah.
* Contoh:
  \[10, 20, 30] → median = 20

  \[10, 20, 30, 100] → median = (20+30)/2 = 25
* **Penggunaan dalam ML:** lebih tahan terhadap outlier, sering digunakan dalam imputasi data (mengganti nilai yang hilang dengan median).

---

### 3. **Mode (Modus)**

* Nilai yang muncul **paling sering**.
* Contoh: \[2, 2, 3, 4] → modus = 2
* **Penggunaan dalam ML:** analisis fitur kategorikal (misalnya mencari kategori yang paling umum).

---

### Perbandingan

| Ukuran | Paling Cocok Digunakan Saat | Kelemahan |
| ------- | -------------------------- | --------- |
| Mean | Data simetris dan tidak memiliki outlier ekstrem | Sensitif terhadap outlier |
| Median | Data menceng (skewed) atau memiliki outlier | Mengabaikan besarnya nilai data |
| Modus | Data kategorikal atau diskret | Mungkin tidak ada atau tidak unik |

---

**Analogi Singkat:**

* **Mean** = "rata-rata gaji seluruh orang."
* **Median** = "gaji orang yang berada tepat di tengah jika semua orang berbaris berdasarkan gajinya."
* **Modus** = "gaji yang paling sering muncul dalam kelompok tersebut."

</details>

> # `Ukuran Penyebaran Data (Measure of Dispersion)`

<details>
<summary>Klik untuk melihat</summary>

### Definisi

* **Penyebaran data (dispersion)** adalah ukuran yang menunjukkan seberapa besar data **bervariasi atau menyebar** di sekitar nilai pusat (mean atau median).
* Membantu kita mengetahui:

  * Apakah data **terkelompok rapat** atau **tersebar luas**?
  * Apakah nilai rata-rata cukup representatif untuk menggambarkan data?

---

### Ukuran Penyebaran Utama

### 1. **Rentang (Range)**

* Rumus:

  $$
  \text{Range} = \text{Nilai maksimum} - \text{Nilai minimum}
  $$
* Contoh: \[10, 20, 30, 40] → Range = 40 – 10 = 30
* **Sederhana tetapi memiliki kelemahan** karena hanya bergantung pada dua nilai ekstrem.

---

### 2. **Rentang Interkuartil (Interquartile Range / IQR)**

* Rumus:

  $$
  IQR = Q_3 - Q_1
  $$

  dengan $Q_1$ = persentil ke-25 dan $Q_3$ = persentil ke-75.
* Contoh: data \[10, 20, 30, 40, 50]

  * $Q_1 = 20$, $Q_3 = 40$ → IQR = 20
* **Penggunaan:** tahan terhadap outlier dan cocok untuk data yang menceng (skewed).

---

### 3. **Varians (Variance)**

* Rumus:

  $$
  \sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}
  $$
* Contoh: \[2, 4, 6]

  * Mean = 4
  * Varians = ((2−4)² + (4−4)² + (6−4)²)/3 = (4+0+4)/3 = 8/3 ≈ 2,67
* **Interpretasi:** semakin besar varians, semakin besar penyebaran data.

---

### 4. **Standar Deviasi (Standard Deviation / SD)**

* Rumus:

  $$
  \sigma = \sqrt{\sigma^2}
  $$
* Contoh: data yang sama \[2, 4, 6] → SD = √2,67 ≈ 1,63
* **Ukuran yang paling banyak digunakan** karena memiliki satuan yang sama dengan data asli.
* **Penggunaan dalam ML:** penskalaan fitur (normalisasi z-score menggunakan mean dan SD).

---

### 5. **Koefisien Variasi (Coefficient of Variation / CV)**

* Rumus:

  $$
  CV = \frac{\sigma}{\bar{x}} \times 100\%
  $$
* Menunjukkan **variasi relatif**, yaitu seberapa besar standar deviasi dibandingkan dengan mean.
* Berguna untuk membandingkan tingkat variasi antara dataset yang memiliki satuan atau skala berbeda.

---

### Tabel Perbandingan

| Ukuran | Rumus | Keterangan |
| ------- | ------- | ---------- |
| Rentang (Range) | Max – Min | Cepat dihitung, tetapi bergantung pada nilai ekstrem |
| IQR | Q3 – Q1 | Tahan terhadap outlier |
| Varians | Rata-rata kuadrat selisih dari mean | Sulit diinterpretasikan karena satuannya kuadrat |
| Standar Deviasi (SD) | √Varians | Paling umum digunakan, satuannya sama dengan data |
| Koefisien Variasi (CV) | SD ÷ Mean × 100 | Membandingkan variasi antar dataset |

---

**Analogi Singkat:**

* Bayangkan tinggi badan para siswa:

  * Jika hampir semua siswa memiliki tinggi sekitar **165 cm**, maka penyebarannya **kecil**.
  * Jika ada yang **150 cm** dan ada yang **190 cm**, maka penyebarannya **besar**.

</details>

> # `Kuantil dan Persentil (Quantiles and Percentiles)`

<details>
<summary>Klik untuk melihat</summary>

### **1. Kuantil (Quantiles)**

* **Definisi**: Kuantil membagi suatu dataset menjadi **bagian-bagian yang sama besar** berdasarkan distribusi data.
* **Bentuk umum**:
  Jika data dibagi menjadi **q bagian yang sama besar**, maka setiap bagian merepresentasikan **1/q bagian** dari keseluruhan data.

### **Jenis Kuantil yang Umum**

* **Kuartil (Quartiles)** (q = 4) → membagi data menjadi **4 bagian yang sama besar**

  * Q1 = titik 25% (kuartil pertama)
  * Q2 = titik 50% (Median)
  * Q3 = titik 75% (kuartil ketiga)
* **Kuintil (Quintiles)** (q = 5) → membagi data menjadi 5 bagian yang sama besar
* **Desil (Deciles)** (q = 10) → membagi data menjadi 10 bagian yang sama besar
* **Persentil (Percentiles)** (q = 100) → membagi data menjadi 100 bagian yang sama besar

**Contoh**:
Jika nilai siswa telah diurutkan,

* Q1 = persentil ke-25 (25% siswa memperoleh nilai di bawah nilai ini)
* Q2 = persentil ke-50 (median)
* Q3 = persentil ke-75

---

### **2. Persentil (Percentiles)**

* **Definisi**: Persentil ke-p adalah nilai yang berada di bawahnya sebanyak **p% dari seluruh observasi**.
* Rentang: persentil ke-0 = nilai minimum, persentil ke-100 = nilai maksimum.
* **Penggunaan umum**:

  * Persentil ke-90 → nilai yang berada di atas 90% data lainnya.
  * Persentil ke-99 → membantu mendeteksi outlier yang sangat ekstrem.

**Contoh**:
Dalam sebuah ujian, jika nilai Anda berada pada **persentil ke-80**, artinya **Anda memperoleh nilai lebih baik daripada 80% siswa lainnya**.

---

### **3. Hubungan antara Kuantil dan Persentil**

* Kuantil adalah konsep umum, sedangkan persentil merupakan **kasus khusus** dari kuantil.
* **Q1 = persentil ke-25**, **Q2 = persentil ke-50**, dan **Q3 = persentil ke-75**.

---

### **4. Penerapan dalam Machine Learning**

* **Deteksi outlier**: menggunakan IQR (Q3 − Q1).

  * Jika suatu titik data < Q1 − 1,5×IQR atau > Q3 + 1,5×IQR → dianggap sebagai outlier.
* **Penskalaan fitur (feature scaling)**: Robust Scaler pada sklearn menggunakan **median dan IQR** (bukan mean dan standar deviasi).
* **Metrik evaluasi**: persentil membantu mengukur **distribusi performa model** (misalnya latensi pada sistem produksi).
* **Menangani data yang menceng (skewed)**: melakukan pengelompokan (binning) berbasis persentil untuk encoding data kategorikal.

---

### **Ringkasan Singkat**

* Kuantil = konsep umum untuk membagi data menjadi beberapa bagian yang sama besar.
* Persentil = pembagian data menjadi 100 bagian yang sama besar (lebih rinci).
* Kuartil (Q1, Q2, Q3) merupakan persentil khusus (25, 50, 75).
* Sangat berguna dalam machine learning untuk **EDA, penskalaan data, deteksi outlier, dan analisis keadilan (fairness analysis)**.

</details>

> # `Ringkasan Lima Angka (Five-Number Summary) dan Boxplot`

<details>
<summary>Klik untuk melihat</summary>

### **Ringkasan Lima Angka (Five-Number Summary)**

Ringkasan lima angka (**five-number summary**) mendeskripsikan suatu dataset menggunakan **lima statistik utama**:

1. **Minimum** → Nilai observasi terkecil.
2. **Q1 (Kuartil Pertama / Persentil ke-25)** → 25% data berada di bawah titik ini.
3. **Median (Q2 / Persentil ke-50)** → Nilai tengah yang membagi dataset menjadi dua bagian sama besar.
4. **Q3 (Kuartil Ketiga / Persentil ke-75)** → 75% data berada di bawah titik ini.
5. **Maksimum** → Nilai observasi terbesar.

---

### **Ringkasan Rumus**

* Kuartil (Q1, Q2, Q3) dihitung dari data yang telah diurutkan.
* **Rentang Interkuartil (IQR) = Q3 − Q1** (digunakan untuk mengukur penyebaran dan mendeteksi outlier).

---

### **Contoh**

Dataset (sudah diurutkan):

`[2, 4, 7, 10, 12, 15, 18, 21, 25]`

* Minimum = 2
* Q1 = 7 (persentil ke-25)
* Median (Q2) = 12
* Q3 = 18 (persentil ke-75)
* Maksimum = 25

**Ringkasan lima angka = (2, 7, 12, 18, 25)**

---

### **Penerapan dalam Machine Learning**

* **Boxplot**: Memvisualisasikan ringkasan lima angka.
* **Deteksi outlier**: Titik data yang berada di luar rentang `[Q1 − 1.5×IQR, Q3 + 1.5×IQR]`.
* **Penskalaan fitur (feature scaling)**: Median dan kuartil lebih **tahan terhadap data menceng (skewed) dan outlier** dibandingkan mean dan standar deviasi.
* **EDA (Exploratory Data Analysis)**: Memberikan ringkasan cepat mengenai bentuk distribusi data (penyebaran dan kemencengan/skewness).

---

**Mnemonik Singkat** → "**Min–Q1–Median–Q3–Maks**"

---

### **Boxplot (Diagram Kotak dan Whisker)**

### **1. Definisi**

**Boxplot** adalah representasi grafis dari distribusi suatu dataset yang menggunakan **ringkasan lima angka**:

* **Minimum**
* **Q1 (Persentil ke-25)**
* **Median (Q2 / Persentil ke-50)**
* **Q3 (Persentil ke-75)**
* **Maksimum**

Boxplot juga menampilkan **outlier** sebagai titik-titik terpisah di luar whisker.

---

### **2. Struktur Boxplot**

* **Kotak (Box)** → Membentang dari Q1 hingga Q3 (Rentang Interkuartil / IQR).
* **Garis di dalam kotak** → Median (Q2).
* **Whisker** → Memanjang dari kotak menuju nilai minimum dan maksimum yang masih berada dalam batas *1,5×IQR*.
* **Outlier** → Titik yang berada di luar whisker, biasanya digambarkan sebagai titik atau bintang.

---

### **3. Contoh**

Jika dataset = `[2, 4, 7, 10, 12, 15, 18, 21, 25]`

* Min = 2
* Q1 = 7
* Median = 12
* Q3 = 18
* Maks = 25

Maka boxplot akan menunjukkan:

* Kotak dari 7 → 18
* Garis pada 12 (median)
* Whisker dari 2 → 25

---

### **4. Penerapan dalam Machine Learning**

* **Deteksi Outlier** → Cara cepat untuk menemukan nilai ekstrem.
* **Bentuk Distribusi** → Menunjukkan kemencengan (skewness), misalnya ketika median lebih dekat ke bagian atas atau bawah kotak.
* **Membandingkan Kelompok** → Boxplot berdampingan memudahkan perbandingan distribusi fitur antar kategori.

  * Contoh: membandingkan **distribusi gaji berdasarkan tingkat pendidikan**.
* **Feature Engineering** → Membantu menentukan apakah data perlu ditransformasi, diskalakan, atau apakah outlier perlu ditangani.

---

### **5. Kelebihan**

* Merangkum distribusi data secara ringkas.
* Cocok untuk **data menceng (skewed)** dan data yang memiliki **outlier**.
* Mudah digunakan untuk membandingkan beberapa dataset secara berdampingan.

---

**Ringkasan:**

Boxplot adalah **visualisasi sekaligus ringkasan statistik** yang menunjukkan penyebaran data, median, dan outlier.

Boxplot membantu menjawab pertanyaan:

* Di mana pusat data berada?
* Seberapa besar penyebaran datanya?
* Apakah terdapat titik data yang tidak biasa (outlier)?

</details>

> # `Grafik untuk Analisis Univariat (Graphs for Univariate Analysis)`

<details>
<summary>Klik untuk melihat</summary>

*Analisis univariat* adalah analisis terhadap **satu variabel** pada satu waktu. Tujuannya adalah untuk memahami distribusi, ukuran pemusatan, dan penyebaran variabel tersebut. Metode visualisasi grafis sangat membantu dalam proses ini.

---

### **1. Histogram**

* **Definisi**: Grafik berbentuk batang di mana data dikelompokkan ke dalam interval-interval (disebut *bin*).
* **Tujuan**: Menampilkan **distribusi frekuensi** dari data kontinu.
* **Contoh dalam ML**: Memvisualisasikan distribusi usia dalam suatu dataset untuk memeriksa apakah distribusinya normal atau menceng (skewed).
* **Catatan**: Membantu mendeteksi kemencengan (skewness), outlier, dan jumlah puncak distribusi (unimodal, bimodal, dan sebagainya).

---

### **2. Diagram Batang (Bar Chart)**

* **Definisi**: Grafik dengan batang persegi panjang yang merepresentasikan frekuensi setiap kategori.
* **Tujuan**: Digunakan untuk **data kategorikal** (nominal atau ordinal).
* **Contoh dalam ML**: Menampilkan jumlah pelanggan berdasarkan jenis kelamin (Laki-laki vs Perempuan).

---

### **3. Diagram Lingkaran (Pie Chart)**

* **Definisi**: Diagram berbentuk lingkaran yang dibagi menjadi beberapa irisan.
* **Tujuan**: Menunjukkan **proporsi** masing-masing kategori.
* **Contoh dalam ML**: Menampilkan proporsi email spam dan non-spam dalam suatu dataset.
* **Catatan**: Tidak direkomendasikan jika jumlah kategori banyak karena diagram batang biasanya lebih mudah dibaca.

---

### **4. Poligon Frekuensi (Frequency Polygon)**

* **Definisi**: Grafik garis yang menghubungkan titik tengah dari setiap bin pada histogram.
* **Tujuan**: Memudahkan perbandingan dua atau lebih distribusi data.
* **Contoh dalam ML**: Membandingkan distribusi nilai ujian antara dua kelas yang berbeda.

---

### **5. Box Plot (Diagram Kotak dan Whisker)**

* **Definisi**: Grafik yang menampilkan **median, kuartil, dan outlier**.
* **Tujuan**: Mengidentifikasi **penyebaran data dan kemencengan (skewness)** serta menyoroti outlier.
* **Contoh dalam ML**: Mendeteksi outlier pada data gaji karyawan.

---

### **6. Stem-and-Leaf Plot (Diagram Batang dan Daun)**

* **Definisi**: Memisahkan angka menjadi bagian **stem** (digit awal) dan **leaf** (digit akhir).
* **Tujuan**: Menampilkan nilai data asli sekaligus distribusinya.
* **Contoh dalam ML**: Mengeksplorasi dataset berukuran kecil (misalnya nilai ujian 30 siswa).

---

### **7. Density Plot (KDE Plot)**

* **Definisi**: Versi histogram yang dihaluskan menggunakan estimasi kepadatan probabilitas (*probability density*).
* **Tujuan**: Membantu memahami **bentuk distribusi probabilitas** suatu variabel.
* **Contoh dalam ML**: Memeriksa apakah suatu fitur dalam dataset mengikuti distribusi Gaussian (Normal).

---

### **Tabel Ringkasan**

| Jenis Grafik | Jenis Data | Tujuan |
| ------------ | ---------- | ------- |
| Histogram | Kontinu | Distribusi frekuensi |
| Diagram Batang (Bar Chart) | Kategorikal | Perbandingan kategori |
| Diagram Lingkaran (Pie Chart) | Kategorikal | Menunjukkan proporsi |
| Poligon Frekuensi | Kontinu | Membandingkan distribusi |
| Box Plot | Kontinu | Penyebaran data dan outlier |
| Stem-and-Leaf Plot | Diskret | Menampilkan nilai asli dan distribusi |
| Density Plot (KDE) | Kontinu | Kurva distribusi yang halus |

---

Dalam machine learning, grafik-grafik ini sering menjadi **langkah pertama dalam EDA (Exploratory Data Analysis)** untuk memahami distribusi fitur sebelum membangun model.

</details>

> # `Grafik untuk Analisis Bivariat (Graphs for Bivariate Analysis)`

<details>
<summary>Klik untuk melihat</summary>
    
*Analisis bivariat* adalah analisis yang mempelajari **hubungan antara dua variabel** secara bersamaan. Variabel tersebut dapat berupa **kategorikal vs kategorikal**, **kategorikal vs numerik**, atau **numerik vs numerik**.

---

### **1. Scatter Plot (Diagram Sebar)**

* **Definisi**: Grafik yang menampilkan titik-titik pada bidang 2D, dengan satu variabel pada sumbu x dan variabel lainnya pada sumbu y.
* **Tujuan**: Menunjukkan **hubungan, korelasi, dan pola** antara dua variabel numerik.
* **Contoh dalam ML**: Hubungan antara luas rumah (sqft) dan harga rumah.
* **Catatan**: Garis tren (*trend line*) dapat ditambahkan untuk menunjukkan hasil regresi.

---

### **2. Line Chart (Diagram Garis)**

* **Definisi**: Grafik yang menghubungkan titik-titik data dengan garis.
* **Tujuan**: Menunjukkan **tren** pada variabel yang bersifat kontinu.
* **Contoh dalam ML**: Memvisualisasikan nilai *loss* terhadap *epoch* selama pelatihan jaringan saraf (*neural network*).

---

### **3. Heatmap (Matriks Korelasi)**

* **Definisi**: Grid berwarna yang menunjukkan nilai korelasi antar beberapa variabel.
* **Tujuan**: Mengidentifikasi **kekuatan dan arah hubungan** antar variabel.
* **Contoh dalam ML**: Memeriksa korelasi antara fitur-fitur dengan variabel target.
* **Catatan**: Korelasi positif biasanya ditunjukkan dengan warna yang lebih gelap atau nilai yang lebih tinggi, sedangkan korelasi negatif dengan warna yang lebih terang atau nilai yang lebih rendah.

---

### **4. Grouped / Clustered Bar Chart (Diagram Batang Berkelompok)**

* **Definisi**: Diagram batang yang dikelompokkan berdasarkan satu variabel kategorikal dan dibandingkan terhadap kategori lainnya.
* **Tujuan**: Membandingkan jumlah atau frekuensi antar kelompok kategori.
* **Contoh dalam ML**: Membandingkan tingkat kelangsungan hidup penumpang Titanic berdasarkan jenis kelamin dan kelas penumpang.

---

### **5. Stacked Bar Chart (Diagram Batang Bertumpuk)**

* **Definisi**: Diagram batang yang terdiri dari beberapa subkategori yang ditumpuk dalam satu batang.
* **Tujuan**: Menunjukkan **komposisi** di dalam setiap kategori.
* **Contoh dalam ML**: Proporsi pelanggan laki-laki dan perempuan pada berbagai kelompok usia.

---

### **6. Side-by-Side Boxplots (Boxplot Berdampingan)**

* **Definisi**: Beberapa boxplot yang ditampilkan berdampingan untuk dibandingkan antar kategori.
* **Tujuan**: Menunjukkan penyebaran variabel numerik pada berbagai kategori.
* **Contoh dalam ML**: Distribusi gaji berdasarkan tingkat pendidikan.

---

### **7. Bubble Chart (Diagram Gelembung)**

* **Definisi**: Pengembangan dari scatter plot di mana **ukuran gelembung** merepresentasikan variabel ketiga.
* **Tujuan**: Menambahkan informasi tambahan pada scatter plot.
* **Contoh dalam ML**: Memvisualisasikan populasi (ukuran gelembung), PDB (GDP), dan tingkat literasi suatu negara secara bersamaan.

---

### **8. Violin Plot**

* **Definisi**: Kombinasi antara boxplot dan density plot.
* **Tujuan**: Menunjukkan distribusi data numerik pada berbagai kategori.
* **Contoh dalam ML**: Distribusi nilai ujian pada beberapa sekolah yang berbeda.
