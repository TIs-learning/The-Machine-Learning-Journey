## Load Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

## Load Dataset

```python
df = pd.read_csv(r"C:\Belajar Pemrograman\Artificial Intelligence\Datasets\Regression\train.csv")
dataset = pd.DataFrame(df)
dfnumeric = df.select_dtypes(include=['number'])
dfnumeric.corr()['SalePrice'].sort_values(ascending=False)
```

## Prepocessing Dataset
Bagi dataset menjadi 4 bagian: Xtrain, Ytrain, Xtest, Ytest

```python
X = np.array(df[['OverallQual',
                  'GrLivArea', 
                  'GarageCars', 
                  'GarageArea', 
                  'TotalBsmtSF', 
                  '1stFlrSF',
                  'FullBath',
                  'TotRmsAbvGrd',
                  'YearBuilt',
                  'YearRemodAdd']])
y = np.array(df['SalePrice'])

#Xtest = np.array
#print(df['OverallQual'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
```

```python
y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)

print(f'R-squared: {r_squared}')
```

## Cek Outlier setiap feature
info: (gagal cok, terakhir sampai yang diatas)
info next paling bawah

```python
Q1 = df['TotRmsAbvGrd'].quantile(0.25)
Q3 = df['TotRmsAbvGrd'].quantile(0.75)

IQR = Q3 - Q1
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR
outlier = df[(df['TotRmsAbvGrd'] < batas_bawah) | (df['TotRmsAbvGrd'] > batas_atas)]
outlier
```

```python
OveralQmedian = df['TotalBsmtSF'].median()
df.loc[(df['OverallQual'] < batas_bawah) | (df['OverallQual'] > batas_atas), 'OverallQual'] = OveralQmedian
#df['OverallQual'] = df['OverallQual'].replace(6.0993150684931505, 6)
df['OverallQual'].value_counts()
# Sukses mengganti nilai kolom OverallQual
```
