# Introduction
(Lasso, Ridge, ElasticNet)

## 1. Ridge Regression
Ridge ini cara kerjanya ngasih 'hukuman' (penalti) ke model kalau ada fitur atau variabel yang bobot pengaruhnya terlalu besar. Hukuman ini dihitung berdasarkan nilai kuadrat dari koefisien fitur tersebut.

Efek utamanya: Ridge bakal menekan atau mengecilkan nilai koefisien fitur-fitur yang kurang penting sampai angkanya sangat kecil mendekati nol, tapi tidak akan pernah sampai benar-benar jadi angka nol mutlak. Artinya, mau se-nggak penting apapun fiturnya, semuanya bakal tetap ikut dipakai dan diperhitungkan sama model ini.

## 2. Lasso Regression

Cara kerja Lasso ini sebenernya mirip banget sama Ridge, dia sama-sama ngasih hukuman ke model. Bedanya, kalau Ridge ngitung hukumannya pakai nilai kuadrat, Lasso ini ngitung hukumannya pakai nilai absolut (nilai mutlak tanpa peduli plus atau minus) dari koefisien fiturnya.

Nah, karena cara ngitung hukumannya beda, efek akhirnya juga beda drastis nih, Pin. Lasso ini karakternya jauh lebih "kejam". Kalau dia ngerasa ada fitur yang kurang penting atau nggak relevan buat nebak target, dia bakal menekan nilai koefisien fitur itu sampai benar-benar jadi angka nol (0).

Artinya, fitur yang nilainya nol itu otomatis dibuang dan nggak dipakai lagi sama model. Makanya, teknik Lasso ini sering banget diandalkan secara otomatis buat feature selection (seleksi fitur).

## 3. ElasticNet Regularization

Singkatnya, ElasticNet ini adalah gabungan atau hybrid dari Ridge dan Lasso. Jadi, dia langsung pakai dua jenis hukuman sekaligus dalam satu model (hukuman kuadrat dari Ridge dan hukuman absolut dari Lasso).

Efeknya gimana? Dia ngambil kelebihan dari keduanya. ElasticNet bisa ngebuang fitur yang nggak penting sampai jadi nol persis kayak Lasso, tapi dia juga lebih stabil dan nggak asal buang kalau kebetulan ada beberapa fitur yang saling berkaitan erat (korelasi tinggi) berkat bantuan sifatnya Ridge.
