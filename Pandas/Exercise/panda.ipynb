```python
import pandas as pd
# membuat dataframe
data = {
    'buah':['apel','mangga','rambutan'],
    'jumlah_buah':[10, 7, 32]
}
df = pd.DataFrame(data)

# Menseleksi dataframe
df[df['jumlah_buah'] > 15]

# Membuat kolom baru
df['harga buah'] = [2000,3000,4000]
print(f'Harga total:\n')
df['Harga Total Buah'] = df['jumlah_buah'] * df['harga buah'] 

# Mentotalkan harga
aset_total = df['Harga Total Buah'].sum()
df
print(f'Aset total adalah: {aset_total}')

# Melihat statistik 
df.describe()

# menyimpan data
df.to_csv('Laporan buah_1.csv', index=False)
```

```python
# Membuat dataframe
data_nilai = {
    'Siswa': ['Ali', 'Budi', 'Citra', 'Deni'],
    'Math': [80, 90, None, 70],
    'Indo': [85, None, 90, 75]
}
df_nilai = pd.DataFrame(data_nilai)
df_nilai.isna() # untuk mengetahui dimana data yang kosong
rataMath = df_nilai['Math'].mean() # mencari tahu rata rata untuk di isi di NaN
rataIndo = df_nilai['Indo'].mean() # mencari tahu rata rata untuk di isi di NaN
df_nilai['Math'] = df_nilai['Math'].fillna(rataMath) # Mengisi NaN dengan rata rata
df_nilai['Indo'] = df_nilai['Indo'].fillna(rataIndo) # Mengisi NaN dengan rata rata
df_nilai 
```

```python
import matplotlib.pyplot as plt
# Membuat dataframe
data_kelas = {
    'Nama': ['Ali', 'Budi', 'Citra', 'Deni', 'Eka', 'Fani'],
    'Kelas': ['10A', '10A', '10B', '10B', '10A', '10B'],
    'Nilai': [85, 90, 75, 85, 95, 70]
}
df_sekolah = pd.DataFrame(data_kelas)
df_sekolah.groupby("Kelas")['Nilai'].mean()

# Kita simpen dulu hasil hitungan rata-rata tadi ke variabel
rata_kelas = df_sekolah.groupby('Kelas')['Nilai'].mean()

# Sekarang kita suruh dia gambar
rata_kelas.plot(kind='bar', color=['blue', 'orange'])
plt.title('Rata-rata Nilai per Kelas')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_toko = {
    'Produk': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headset', 'Webcam', 'Laptop', 'Mouse'],
    'Kategori': ['Komputer', 'Aksesoris', 'Aksesoris', 'Komputer', 'Aksesoris', 'Aksesoris', 'Komputer', 'Aksesoris'],
    'Harga': [15000000, 200000, None, 3000000, 500000, 800000, 15000000, 200000],
    'Terjual': [5, 20, 15, None, 10, 8, 5, 20]
}

df_toko = pd.DataFrame(data_toko)
print("Data Berhasil Dibuat!")
#df_toko.describe()
medianHarga = df_toko['Harga'].median()
medianTerjual = df_toko['Terjual'].median()
print(f'Median harga adalah: {medianHarga}')
print(f'Median Terjual adalah: {medianTerjual}')
df_toko["Harga"] = df_toko["Harga"].fillna(medianHarga)
df_toko["Terjual"] = df_toko["Terjual"].fillna(medianTerjual)
df_new = df_toko.drop_duplicates()
df_new['Total Keuntungan'] = df_new['Harga'] * df_new['Terjual']
df_new['Total Keuntungan'].sort_values(ascending=False)
df_new.groupby('Kategori')['Total Keuntungan'].sum()

# Ambil data omzet per kategori
omzet_kategori = df_new.groupby('Kategori')['Total Keuntungan'].sum()

# Bikin diagram lingkaran (Pie Chart)
#omzet_kategori.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
#plt.title('Persentase Kontribusi Omzet per Kategori')
#plt.ylabel('') # Biar tulisan 'Total' di samping gak ganggu
#plt.show()
df_new.groupby('Kategori')['Harga'].describe()
```

```python
import pandas as pd

# Tabel 1: Data Pesanan (Siapa beli Menu ID apa)
df_pesanan = pd.DataFrame({
    'ID_Pesanan': [101, 102, 103, 104],
    'ID_Menu': [1, 3, 2, 1],
    'Jumlah': [2, 1, 3, 5]
})

# Tabel 2: Daftar Harga & Nama Menu
df_menu = pd.DataFrame({
    'ID_Menu': [1, 2, 3],
    'Nama_Menu': ['Ayam Bakar', 'Es Teh', 'Nasi Goreng'],
    'Harga': [20000, 5000, 15000]
})
df_restoran2 = pd.merge(df_pesanan, df_menu, on='ID_Menu', how='left')
df_restoran2['Subtotal'] = df_restoran2['Jumlah'] * df_restoran2['Harga']
total = df_restoran2['Subtotal'].sum()
print("Tabel Pesanan:")
print(df_pesanan)
print("\nTabel Menu:")
print(df_menu)
print("\nTabel Restoran_2:")
print(df_restoran2)
df_restoran2.sort_values(by='Subtotal', ascending=False)
total
df_restoran2.to_csv('LaporanResto.csv', index=False)
```
