```python
import pandas as pd

# Dataset 1: Data Karyawan
df_karyawan = pd.DataFrame({
    'ID_Karyawan': [1, 2, 3, 4],
    'Nama': ['Andi', 'Budi', 'Citra', 'Dewi'],
    'ID_Departemen': [101, 102, 101, 104]
})

# Dataset 2: Data Departemen
df_departemen = pd.DataFrame({
    'ID_Departemen': [101, 102, 103],
    'Nama_Departemen': ['HR', 'IT', 'Finance']
})

# Coba gabungkan di sini!
# Contoh: pd.merge(df_karyawan, df_departemen, on='ID_Departemen', how='inner')
```

```python
# merge untuk right join
data = pd.merge(
    df_karyawan, # data a
    df_departemen, # data b
    on = 'ID_Departemen', # kolom dimiliki / sama sama punya 
    how= 'right' # yang menjadi acuan
    # kita juga bisa menambahkan lsuffix='_ret' / rsuffix='_iu' untuk menambahkan nama nama kolom
)
data
```
