apa itu data understanding and profilling? proses memahami data seperti struktur, isi, kualitas data sebelum dilakukannya analisis.

struktur data: jumlah baris kolom, nama kolom, tipe data setiap kolom

ringkasan data: statistik dasar, distribusi nilai, frekuensi data

missing value analysis: kolom yang kosong, proporsi missing value, pola missing data

duplikat baris: data duplikat baris, duplikat berdasarkan kolom tertentu

value distribution: Nilai unik, frekuensi kategori, data tidak wajar

data consistency: format data konsisten, nilai di luar aturan bisnis, relasi antar kolom masuk akal
