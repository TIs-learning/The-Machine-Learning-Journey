Apa itu data wrangling? proses kita melakukan perubahan terhadap struktur data dengan melakukan import(load) data, penggabungan data, data cleaning, data analyst, dsb.

struktur data di pandas ada Series dan juga Dataframe, series itu seperti kolom pada python yang hanya mempunya 1 dimensi [...] yang berisi 1 tipe data dengan index sedangkan dataset terdiri dari berbagai tipe data series beberapa dataseries tersebut juga disebut sebagai dataframe.

1. Data ingestion adalah proses untuk meload data yang ingin kita masukkan untuk membuat suatu dataframe dari berbagai sumber, di pandas kita bisa dari database atau file dengan tipe data json, excel, csv, parquet = pd.read_csv (baca dokumentasi)

2. data integration adalah proses menggabungkan beberapa dataset untuk membuat suatu dataframe yang utuh, kita bisa menggunakan concat u/ via index or kolom, merge antar dataser, join via index dengan tipe join inner yang cocok aja, left bagian kiri, right bagian kanan, outer semua data.
