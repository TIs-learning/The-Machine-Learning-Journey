```python
# Pandas 02 | Pemilihan baris (rows selection) pada Data Frame | Python Pandas | Data Science
# membuat persiapan dataframe
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), columns=cols)
df.add_prefix('kolom_') # oh gua baru ngeh bedanya pre and suffix depan belakang
# selection dengan operator logika '|' nah itu pengandaian or 
df[(df['A'] == 1) | (df['A'] == 3)] #hanya memilih kolom a bernilai 1 atau 3 lalu ditampilkan secara baris
# selection dengan fungsi isin(), ini untuk menggantikan fungsi yang sebelumnya |
df[df['A'].isin([1, 3])]
# mengenal operator negasi ~, yang tidak sama dengan 1 atau 3
df[~df['A'].isin([1, 3])]
```

```python
# andas 33 | Nilai Kumulatif pada Pandas Data Frame | Python Pandas | Data Science | Bahasa Indonesia
# 1. Membuat dataframe
d = {'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'Joni'], 
     'goal':[2, 1, 3, 1, 1, 2, 2, 3]}
df = pd.DataFrame(d)
df
# 2. Menampilkan nilai kumulatif (cummulative)
df['goal'].cumsum().to_frame()
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
df['cummax'] = df['goal'].cummax()
df
df['cummin'] = df['goal'].cummin()
df
df['cumprod'] = df['goal'].cumprod()
df
# ini buat lihat nilai kumulatif, paling besar/kecil, trus hasil perkalian kumulatig
```

```python
# Pandas 34 | Mapping Data Frame dengan applymap | Python Pandas | Data Science | Bahasa Indonesia
# 1. Membuat dataframe
df = pd.DataFrame({'jenis_kelamin':['Pria', 'Wanita', 'lelaki', 'Lelaki', 'perempuan'], 
                   'usia':[23, 21, 24, 22, 21], 
                   'shift':['pagi', 'siang', 'Malam', 'Siang', 'pagi']})
df
# 2. Mapping pada Data Frame dengan applymap()
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
# 3. applymap() dengan dictionary
mapping = {'pria':'L',
           'lelaki':'L',
           'wanita':'P',
           'perempuan':'P',
           'pagi':1,
           'siang':2,
           'malam':3}
# ini kayak membuat standarisasi semuanya semisal ada yang mau huruf besar/kecil semua
df.applymap(mapping.get)
df[['jenis_kelamin', 'shift']] = df[['jenis_kelamin', 'shift']].applymap(mapping.get)

df
```

```python
# Pandas 35 | Memadukan fungsi agregasi dengan transform | Python Pandas | Belajar Data Science
# 1. Membuat dataframe
d = {'no_nota':[1, 1, 1, 2, 2, 3, 4, 5], 
     'kopi': ['latte', 'cappuccino', 'espresso', 'latte', 'espresso', 'cappuccino', 'latte', 'espresso'],
     'harga':[50, 60, 80, 150, 120, 60, 100, 40]}

df = pd.DataFrame(d)
df
# 2. Menghitung total harga untuk tiap nomor nota
df.groupby('no_nota')['harga'].sum().to_frame()
df['total_harga'] = df.groupby('no_nota')['harga'].transform(sum)
df
# 3. Menghitung total omset untuk tiap jenis kopi yang terjual
df.groupby('kopi')['harga'].sum().to_frame()
df['total_omset'] = df.groupby('kopi')['harga'].transform(sum)
df
```
