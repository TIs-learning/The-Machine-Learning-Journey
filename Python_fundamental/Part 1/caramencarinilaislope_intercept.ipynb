## 📐 Regresi Linear Sederhana

### 🎯 Persamaan Dasar

$$
y = wx + b
$$

---

### 📊 Cara Menghitung Slope ($w$)

$$
w = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

**Makna:**

* Mengukur kekuatan hubungan linear antara $x$ dan $y$
* Jika:

  * $w > 0$ → hubungan positif (naik)
  * $w < 0$ → hubungan negatif (turun)

---

### 📊 Cara Menghitung Intercept ($b$)

$$
b = \bar{y} - w\bar{x}
$$

**Makna:**

* Nilai $y$ saat $x = 0$
* Menggeser garis ke atas atau ke bawah

---

## 🧠 Keterangan Notasi

* $x_i$ = nilai fitur ke-$i$
* $y_i$ = nilai target ke-$i$
* $\bar{x}$ = rata-rata dari seluruh nilai $x$
* $\bar{y}$ = rata-rata dari seluruh nilai $y$
* $w$ = slope (kemiringan garis)
* $b$ = intercept (titik potong sumbu $y$)
* $\sum$ = penjumlahan seluruh data

---

## 🎯 Intuisi Singkat

> Model mencari nilai $w$ dan $b$ agar garis yang terbentuk **paling mendekati semua titik data** (meminimalkan error).

---

```python
import numpy as np

# Data contoh
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Hitung mean
x_mean = np.mean(x)
y_mean = np.mean(y)

# Hitung slope (w)
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

w = numerator / denominator

# Hitung intercept (b)
b = y_mean - w * x_mean

print("Slope (w):", w)
print("Intercept (b):", b)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Data harus 2D untuk X
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

print("Slope (coef_):", model.coef_)
print("Intercept:", model.intercept_)
```
