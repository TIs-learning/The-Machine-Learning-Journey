```python
import numpy as np

# Data sederhana
x = np.array([1, 2, 3])
y = np.array([2, 3, 5])
```

```python
w = 0.0
b = 0.0

learning_rate = 0.01
epochs = 30
```

```python
for epoch in range(epochs):
    y_pred = w * x + b  # prediksi

    error = y - y_pred  # selisih (y - ŷ)

    # Loss (SSE)
    loss = np.sum(error ** 2)

    # hitung gradient
    dw = -2 * np.sum(x * error)
    db = -2 * np.sum(error)

    # update parameter
    w = w - learning_rate * dw
    b = b - learning_rate * db

    print(f"Epoch {epoch}: w={w:.4f}, b={b:.4f}, Loss={loss:.4f}")
```
