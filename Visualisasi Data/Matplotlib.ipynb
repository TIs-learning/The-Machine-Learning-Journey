```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Dasar Matplotlib

```python
x = np.array([5002, 2004, 2005, 2006]) # Definisikan variabel ke 1
y = np.array([15, 25, 34, 67]) # Definisikan variabel ke 2
plt.plot(x, y) # argumen ke 1 untuk membentuk grafik secraa horizontal, ke 2 untuk membentuk secara vertikal
plt.show() # Memanggil untuk dikeluarkan
```

## 2. Kustomisasi Matplotlib

```python
# ms(ukuran) = markersize, mfc(warna marker) = markerfacecolor, mec(warna tepi) = markeredgecolor, color(warna garis) 
# linestyle(tipe garis) = dashed/none/dotted/dashdot/solid, linewidth(tebal)

x = np.array([2007, 2008, 2009, 2010]) # Definisikan variabel ke 1
y1 = np.array([15, 25, 34, 67]) # Definisikan variabel ke 2
y2 = np.array([23, 27, 30, 38])
y3 = np.array([12, 14, 18, 21])

lineStyle = dict(marker=".",
                 ms='10',
                 mfc="#499a6c",
                 mec="#cdae59",
                 linestyle='solid',
                 linewidth='2')

# argumen ke 1 untuk membentuk grafik secraa horizontal, ke 2 untuk membentuk secara vertikal
plt.plot(x, y1, color='#1212c4', **lineStyle)
plt.plot(x, y2, color="#ff1a48", **lineStyle) 
plt.plot(x, y3, color="#ff8d02", **lineStyle) 

plt.show() # Memanggil untuk dikeluarkan
```

## Label

```python
x = np.array([2007, 2008, 2009, 2010]) 
y1 = np.array([15, 25, 34, 67]) 
y2 = np.array([23, 27, 30, 38])
y3 = np.array([12, 14, 18, 21])

plt.title('angka kelahiran', fontsize=20, 
                             family='Arial', 
                             fontweight='bold',
                             color='#2d4cfc')

plt.xlabel("Year", fontsize=15,
                   family='Arial', 
                   fontweight='bold',
                   color='#2d4cfc')

plt.ylabel("Angka Kematian", fontsize=15,
                             family='Arial', 
                             fontweight='bold',
                             color='#2d4cfc')

plt.tick_params(axis='both', # x/y/both
                colors ='#2d4cfc') 

plt.xticks(x)

lineStyle = dict(marker=".",
                 ms='10',
                 mfc="#499a6c",
                 mec="#cdae59",
                 linestyle='solid',
                 linewidth='2')

plt.plot(x, y1, color='#1212c4', **lineStyle)
plt.plot(x, y2, color="#ff1a48", **lineStyle) 
plt.plot(x, y3, color="#ff8d02", **lineStyle) 

plt.show()
```

## Grid Lines

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid(axis='x', # x/y/(kosong)
         linewidth=2,
         color='lightgray',
         linestyle='dashed')

plt.plot(x, y)
plt.show()
```

## Bar Chart (Diagram Batang)

```python
categories = np.array(['Grains', 'fruit', 'vegetables', 'proteins', 'Dairy', 'sweets'])
values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, values, color='skyblue')
# plt.barh kalo mau vertikal

plt.title('Daily Cossumption')
plt.xlabel("food")
plt.ylabel('Quantity')

plt.show()

```

## Pie Chart (Diagram Lingkaran)

```python
categories = np.array(['Freshmen', 'Sophomers', 'Juniors', 'Seniors'])
Values = np.array([300, 250, 270, 225])
colors = ['red', 'yellow', 'blue', 'green']

plt.title('Data Orang Mancing')

plt.pie(Values, labels=categories,
                autopct='%1.2f%%', 
                colors=colors,
                explode=[0, 0, 0, 0.3], # efek memecah 0.1....
                shadow=True, # efek bayangn
                startangle=90) # efek diputar
plt.show()
```

## Scatter Graph (Diagram Sebar)

```python
x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]) # Hours study
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]) # grades

x2 = np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8]) # Hours study
y2 = np.array([50, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97]) # grades

plt.scatter(x1, y1, color='skyblue',
                  alpha=0.5, # Tingkat Transparan
                  s = 200, # s = size
                  label='Class A') 

plt.scatter(x2, y2, color='red',
                  alpha=0.5, 
                  s = 200,
                  label='Class B') 

plt.title('Correlation Between Hours Study Student Vs Grades')
plt.xlabel('Hours Study')
plt.ylabel('Grades')

plt.legend() # supaya ada legenda di atas kiri
plt.show()
```

## Histogram 

```python
scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100) # Menetapkan Batasan nilai bawah dan atas

plt.hist(scores, bins=10, # Banyaknya diagram yang ditampilkan
                 color='lightgreen',
                 edgecolor='black') 

plt.title('Exam Scores')
plt.xlabel('Score')
plt.ylabel('Number of Students')

plt.show()
```

## Subplot

```python
# Figure = Keseluruhan Canvas
# Ax = Canvas tersendiri (Subplot)

x = np.array([1, 2, 3, 4, 5])

figure, axes =plt.subplots(2, 2) # Baris dan kolom 

axes[0, 0].plot(x, x*2, color='red')
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x, x**2, color='blue')
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x, x**3, color='green')
axes[1, 0].set_title("x**3")

axes[1, 1].plot(x, x**4, color='purple')
axes[1, 1].set_title("x**4")

#axes[...] (Grafik tujuan).jenis grafik

plt.tight_layout() # ini supaya ga tabrakan bawah dengan atas
plt.show()
```

## Matplotlib + Pandas

```python
import pandas as pd
df = pd.read_csv('pokemon.csv')

type_count = df['Type1'].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, color='red',
         edgecolor='black')

plt.title("Number of Pokemon by Primary Type")
plt.xlabel('Count')
plt.ylabel('type')
plt.tight_layout()
plt.show()
```
